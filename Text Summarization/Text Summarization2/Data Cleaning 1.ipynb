```python
import pandas as pd
```

```python
Data=pd.read_csv('news_summary.csv',encoding='latin-1')
```

```python
def nullMng(Data,Null_values='mean',ax=None):
    if ax==None:
        for i in range(Data.shape[1]):
            if Null_values=='drop':
                    Data.drop(Data[Data[Data.columns[i]].isnull()].index, axis=0, inplace=True)
                    Data.reset_index(drop=True,inplace=True)
            elif Data[Data.columns[i]].dtype=='O':
                Data[Data.columns[i]][Data[Data.columns[i]].isnull()]='012789'
            else:
                if Null_values=='mean':
                    Data[Data.columns[i]][Data[Data.columns[i]].isnull()]=Data[Data.columns[i]].mean()
                elif Null_values=='mode':
                    Data[Data.columns[i]][Data[Data.columns[i]].isnull()]=Data[Data.columns[i]].mode()
                elif Null_values=='median':
                    Data[Data.columns[i]][Data[Data.columns[i]].isnull()]=Data[Data.columns[i]].median()
                else:
                    Data[Data.columns[i]][Data[Data.columns[i]].isnull()]=Null_values
    else:
        if Null_values=='drop':
                Data.drop(Data[Data[Data.columns[ax]].isnull()].index, axis=0, inplace=True)
                Data.reset_index(drop=True,inplace=True)
        elif Data[Data.columns[ax]].dtype=='O':
            Data[Data.columns[ax]][Data[Data.columns[ax]].isnull()]='012789'
        else:
            if Null_values=='mean':
                Data[Data.columns[ax]][Data[Data.columns[ax]].isnull()]=Data[Data.columns[ax]].mean()
            elif Null_values=='mode':
                Data[Data.columns[ax]][Data[Data.columns[ax]].isnull()]=Data[Data.columns[ax]].mode()
            elif Null_values=='median':
                Data[Data.columns[ax]][Data[Data.columns[ax]].isnull()]=Data[Data.columns[ax]].median()
            else:
                Data[Data.columns[ax]][Data[Data.columns[ax]].isnull()]=Null_values
    return Data
```
